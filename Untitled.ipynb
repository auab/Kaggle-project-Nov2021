## 1. Import libaries and datasets

```python
# Libraries
import pandas as pd
from boruta import BorutaPy
from sklearn.linear_model import SGDClassifier
# Datasets
train_dataset = pd.read_csv('train.csv')
test_dataset = pd.read_csv('test.csv')
# Separate features from values
X_train, y_train = train_dataset.iloc[:,1:-1],train_dataset['target']
X_test = test_dataset.iloc[:,1:]
```

## 2. Treat the data

First thing to do is a feature selection, this is done here with the Boruta-SHAP algorithm. More about in 

## 2. Train the model

```python
sgd_clf = SGDClassifier(loss='log',random_state=42)
sgd_clf.fit(X_train, y_train)
```

## 3. Predict the target value

```python
y_predict = sgd_clf.predict_proba(X_test)
y_proba_1 = [y[1] for y in y_predict]
```

## 4. Save result

```python
y_predict_df = pd.DataFrame(y_proba_1,columns=['target'])
y_csv_format = pd.concat([test_dataset['id'],y_predict_df],axis=1)
y_csv_format.to_csv('y_predicted.csv',index=False)
```
